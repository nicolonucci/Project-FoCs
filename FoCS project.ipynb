# ESERCIZIO 14
### Read all json files with the video categories

```python
json_files = glob.glob(os.path.join(path, "*_category_id.json"))
categories_list = []

for f in json_files:
    country = os.path.basename(f)[:2].upper()
    
    with open(f, "r", encoding="utf-8") as file:
        data = json.load(file)

    for item in data["items"]:
        categories_list.append({"category_id": int(item["id"]),
                                "category_title": item["snippet"]["title"],
                                "country": country,
                                "assignable": item["snippet"].get("assignable", False)})

categories_df = pd.DataFrame(categories_list)
categories_df
```

### For each country, determine how many videos have a category that is not assignable.

```python
merged = full_df.merge(categories_df,how="left",
                       left_on=["category_id", "country"],
                       right_on=["category_id", "country"])
not_assigable = merged[merged['assignable'] == False]
not_assignable_per_country = (not_assigable.groupby("country")["video_id"]
                                            .nunique()
                                            .reset_index(name="not_assignable_videos"))
not_assignable_per_country
```
